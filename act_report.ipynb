##### Looking at how often WeRateDogs posted over the years shows a clear pattern. The number of tweets went up a lot from 2015 to 2016, which points to a time when the account grew and posted more. ### 4- Is there a relation between the length of tweet text and the number of favorites and retweets?


 



##### In measuring tweet lengths against engagement metrics, our exploration finds the following: while there is a positive relationship, weak correlation exists between the length of a tweet and its retweet count, with no compelling evidence to show that longer tweets convert to more shares; rather, there is a stronger positive relationship between tweet lengths and the number of favorites. This may suggest that although the length of the tweet may not affect retweets, it is the longer tweets that get favorited. This could mean that higher details or descriptive tweets are much more appealing to the audience. However, overall preference tends toward short Tweets, given the fact that retweeting was very weakly correlated.
